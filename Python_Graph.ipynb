# 3차 함수 ;f(x) = (x-2)x(x+2)

```python
def f(x):
    y = (x-2)*x*(x+2)
    return y

print(f'x = 1, y = {f(1)}')
print(f'x = 2, y = {f(2)}')
print(f'x = 3, y = {f(3)}')

res = f(np.array([1,2,3])) # list -> np.ndarray
x = np.arange(-3 , 3, 0.5)
x = np.linspace(-3,3,50)
res = f(x)
print(f'x = {x}\ny = {res} \ntype : {type(res)}')

plt.figure(figsize=(5,5))
plt.plot(x,res,'pink',linewidth=2,label = '$y =  (x-2)*x*(x+2)$')
plt.ylim(-30,30)
plt.ylabel('y')
plt.xlim(-4,4)
plt.xlabel('x')
plt.grid(True)
plt.legend(loc='lower right')
plt.show()
```

```python
def f2(x):
    y = (x-2)*x*(x+2)
    return y

def f(x, w):
    y = (x-w)*x*(x+2)
    return y



x = np.linspace(-3,3,50)
res = f2(x)
print(f'x = {x}\ny = {res} \ntype : {type(res)}')

plt.figure(figsize=(5,5))
plt.plot(x,f2(x),'pink',linewidth=2,label = '$y =  (x-2)*x*(x+2)$')
plt.plot(x,f(x,w=4),'orange',linewidth=2,label = '$y =  (x-4)*x*(x+2)$')
plt.plot(x,f(x,w=0),'magenta',linewidth=2,label = '$y =  (x)*x*(x+2)$')
plt.title('$f_2(x)$')
plt.ylim(-30,30)
plt.ylabel('y')
plt.xlim(-4,4)
plt.xlabel('x')
plt.grid(True)
plt.legend(loc='upper left')
plt.show()
```

```python
def f3(x0,x1):
    r = 2*x0**2+x1**2
    ans = r*np.exp(-r)
    return ans
```

```python

#x0,x1에 따라 각각 나오는 f3함수를 계산
xn = 9
x0 = np.linspace(-2,2,xn)
x1 = np.linspace(-2,2,xn)
y = np.zeros((len(x0),len(x1)))
print(f'x0 = {x0}\nx1 = {x1}')

for i0 in range(xn):
    for i1 in range(xn):
        y[i1,i0]=f3(x0[i0],x1[i1])
y = np.round(y,1)
print(f'y = {y}')
plt.figure(figsize=(3.5,3))
plt.gray() #회색
#plt.jet() #색입히기
#plt.pink() #분홍
#plt.bone() #파랑계열

plt.pcolor(y)
#plt.contour(y)

plt.colorbar()
plt.savefig('graph_3.png')
plt.show()
```

## 그래프 표현 - surface

```python
from mpl_toolkits.mplot3d import Axes3D

xx0, xx1 = np.meshgrid(x0,x1)
plt.figure(figsize=(5,3.5))
ax = plt.subplot(1,1,1, projection = '3d')
ax.plot_surface(xx0,xx1,y, rstride = 1, cstride= 1, alpha = 0.3, color = 'blue', edgecolor='black')
ax.set_zticks((0,0.2))
ax.view_init(10,-95)
print(f'xx0 = {xx0}\nxx1 = {xx1}\n{y}')
```

## 그래프 표시 - contour

```python
from mpl_toolkits.mplot3d import Axes3D


#x0,x1에 따라 각각 나오는 f3함수를 계산
xn = 9
x0 = np.linspace(-2,2,xn)
x1 = np.linspace(-2,2,xn)
y = np.zeros((len(x0),len(x1)))
print(f'x0 = {x0}\nx1 = {x1}')

for row in range(xn):
    for col in range(xn):
        y[col,row]=f3(x0[row],x1[col])

cont = plt.contour(xx0, xx1, y, 5, colors ='black')
cont.clabel(fmt='%3.2f', fontsize = 8)
plt.xlabel('$x_0$', fontsize=14)
plt.ylabel('$x_1$', fontsize=14)
#plt.colorbar()
plt.show()
```

```python
def mul(fir,sec):
    res = fir*sec
    return res

colL = ['orange','blue','cyan','magenta']
first = np.linspace(1,9,9)
for i in range(1,5):
    temp = i
    plt.subplot(2,2,i)
    plt.plot(first,mul(first,(temp*2)+1),f'{colL[i-1]}',label = f'${(temp*2)+1}dan$')
    plt.ylim(0,91)
    cnt = 0
    print(f'{int((temp*2)+1)} 단:',end='')
    for k in range(0,len(first)):
        print('\t '+f'{int((temp*2)+1)}x{int(first[k])} = {int(mul(first[k],(temp*2)+1))}',end='')
        cnt += 1
        if cnt == 9:
            print('')
    plt.title(f'{(temp*2)+1}dan')
    plt.grid(True)
    plt.legend(loc='upper left')
plt.show()

plt.figure(figsize=(5,5))
plt.plot(first,mul(first,3),f'{colL[0]}',label = f'$3dan$')
plt.plot(first,mul(first,5),f'{colL[1]}',label = f'$5dan$')
plt.plot(first,mul(first,7),f'{colL[2]}',label = f'$7dan$')
plt.plot(first,mul(first,9),f'{colL[3]}',label = f'$9dan$')
plt.title('Gugudan')
plt.grid(True)
plt.legend(loc='upper left')
plt.show()
```
