# 미분

$f(x) = x^2$ .. x 에 대한 식(1)</br>
$dx_f(x) = 2x$ .. 식(1)에 대해 x에 대한 1차미분

$f(x,w) = x^2 + w^2$ x에 대한 식(2) \
$dx_f(x,w)$ .. 식(2)에 대해 x에 대한 1차 미분\
$dw_f(x,w)$ .. 식(2)에 대해 y에 대한 1차 미분

```python
def f(w0,w1):
    return w0**2 + 2*w0*w1+3     # x^2 + 2xy + 3$\
def df_dw0(w0,w1):       
    return 2*w0 + 2*w1          # 2x + 2y$\
def df_dw1(w0,w1):
    return 2*w0                 # 2x$
    
```

```python
import numpy as np

w_range=2
dw=0.25
w0 = np.arange(-w_range, w_range+dw, dw)
w1 = np.arange(-w_range, w_range+dw, dw)
wn = w0.shape[0]
ww0, ww1 = np.meshgrid(w0,w1)
ff = np.zeros((len(w0),len(w1)))
dff_dw0 = np.zeros((len(w0),len(w1)))
dff_dw1 = np.zeros((len(w0),len(w1)))
```

```python
for i0 in range(wn):
    for i1 in range(wn):
        ff[i1,i0] = f(w0[i0],w1[i1])
        dff_dw0[i1,i0] = df_dw0(w0[i0],w1[i1])
        dff_dw1[i1,i0] = df_dw1(w0[i0],w1[i1])
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(9,4))
plt.subplots_adjust(wspace=0.3)

plt.subplot(1,2,1)
cont = plt.contour(ww0, ww1, ff, 10)
cont.clabel(fmt='%0.2f', fontsize=8)
plt.xticks(range(-w_range, w_range+1,1))
plt.yticks(range(-w_range, w_range+1,1))
plt.xlim(-w_range-0.5,w_range+0.5)
plt.ylim(-w_range-0.5,w_range+0.5)
plt.xlabel('$w_0$', fontsize=14)
plt.ylabel('$w_1$', fontsize=14)

plt.subplot(1,2,2)
plt.quiver(ww0, ww1, dff_dw0, dff_dw1)
plt.xticks(range(-w_range, w_range+1,1))
plt.yticks(range(-w_range, w_range+1,1))
plt.xlim(-w_range-0.5,w_range+0.5)
plt.ylim(-w_range-0.5,w_range+0.5)
plt.xlabel('$w_0$', fontsize=14)
plt.ylabel('$w_1$', fontsize=14)
```

### 함수 f(w)의 w에 대한 미분
$$
\frac{df(w)}{dw},
\frac{d}{dw}f(w), 
f'(w)
$$

### 함수 $f(w)=w^2$의 w에 대한 미분
$$
\frac{d}{dw}f(w) = \frac{d}{dw}w^2 = 2w
$$

### 함수 $f(w)=w^n$의 w에 대한 미분
$$
\frac{d}{dw}f(w) = \frac{d}{dw}w^n = nw^{n-1}
$$

## 중첩된 함수의 미분 : 연쇄법칙
### 대입식 미분
$f(w) = g(w)^2$<br>
$g(w) = aw+b$<br>
$f(w) = (aw+b)^2 = a^2w^2 + 2abw + b^2$<br>
$\frac{d}{dw}f(w) = 2a^2w + 2ab$<br>

### 연쇄법칙 미분
$$
\frac{df}{dw} = \frac{df}{dg}\cdot\frac{dg}{dw}
$$

$\frac{df}{dg}$는 f를 g로 미분한다.
ex) $
\frac{df}{dg} = \frac{d}{dg}g^2=2g
$
<br>
$\frac{dg}{dw}$는 g를 w로 미분한다.
ex) $
\frac{dg}{dw} = \frac{d}{dw}(aw+b)=a
$
<br>
<br>
$
\frac{df}{dw} = \frac{df}{dg}\cdot\frac{dg}{dw} = 2ga = 2(aw+b)a = 2a^2w+2ab
$


### 연쇄법칙 미분...확장
$$
\frac{df}{dw} = \frac{df}{dg}\cdot\frac{dg}{dh}\cdot\frac{dh}{dw}
$$

## non-linear activations

```python
import copy
def binary_step(_x):
    x = copy.deepcopy(_x)
    for i,v in enumerate(x): 
        if v>0 : x[i]=1.0
        if v<0 : x[i]=0.0
    return x
def sigmoid_function(x):
    z = (1/(1 + np.exp(-x)))
    return z    
def tanh_function(x):
    z = (2/(1 + np.exp(-2*x))) -1
    return z
def relu_function(_x):
    x = copy.deepcopy(_x)
    for i,v in enumerate(x): 
        if v>0 : x[i]=x[i]
        if v<0 : x[i]=0.0
    return x

def leaky_relu_function(_x):
    x = copy.deepcopy(_x)
    for i,v in enumerate(x): 
        if v>0 : x[i]=0.01*x[i]
        if v<0 : x[i]=0.0
    return x
def elu_function(x, a):
    if x<0:
        return a*(np.exp(x)-1)
    else:
        return x    
def swish_function(x):
    return x/(1-np.exp(-x))
def softmax_function(x):
    z = np.exp(x)
    z_ = z/z.sum()
    return z_
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10,10,100)
plt.figure(figsize=(5,5))
plt.plot(x,binary_step(x),'gray', linestyle='--', linewidth=1, label='binary_step')
plt.plot(x,sigmoid_function(x),'black', linewidth=1, label='sigmoid')
plt.plot(x,tanh_function(x),'green', linewidth=1, label='tanh')
plt.plot(x,relu_function(x),'orange', linewidth=1, label='relu')
plt.plot(x,leaky_relu_function(x),'red', linewidth=1, label='leaky_relu')
plt.plot(x,elu_function(x,1),'blue', linewidth=1, label='elu')
plt.xlim(-10,10)
plt.ylim(-1,1.2)
plt.grid(True)
plt.legend(loc='lower right')
```
